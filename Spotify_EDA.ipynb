```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


import warnings
warnings.filterwarnings('ignore')

df = pd.read_csv('spotify.csv')
df
```

```python
# Q1. Read the dataframe, check null value if present then do the needful, check duplicate row , if present then do the needful.


# Check for null values
print(df.isnull().sum())

# If null values are present, you can choose to remove rows with null values or fill them with appropriate values.
# Example: Remove rows with null values in any column.
# df = df.dropna()

# Example: Fill null values in the 'Popularity' column with the mean.
# df['Popularity'].fillna(df['Popularity'].mean(), inplace=True)


# Check for duplicate rows
duplicate_rows = df[df.duplicated()]
print(f"Number of duplicate rows: {len(duplicate_rows)}")
print("Duplicate rows:")
duplicate_rows

# If duplicate rows are present, you can choose to remove them.
# df.drop_duplicates(inplace=True)
```

```python
# Q2.What is the distribution of popularity among the tracks in the dataset? Visualize it using a histogram.


# Create the histogram
plt.hist(df['Popularity'], bins=20, edgecolor='black')  # Adjust bins as needed

# Add labels and title
plt.xlabel('Popularity')
plt.ylabel('Frequency')
plt.title('Distribution of Track Popularity')

# Show the plot
plt.show()
```

```python
# Q3.Is there any relationship between the popularity and the duration of tracks? Explore this using a scatter plot.


import plotly.express as px

# Create the scatter plot
fig = px.scatter(df, x="Popularity", y="Duration (ms)",
                 title="Relationship between Popularity and Duration of Tracks")

# Show the plot
fig.show()
```

```python
# Q4. Which artist has the highest number of tracks in the dataset? Display the count of tracks for each artist using a countplot.


# Calculate the number of tracks per artist
artist_counts = df['Artist'].value_counts()

# Find the artist with the highest number of tracks
artist_with_most_tracks = artist_counts.index[0]
count_of_most_tracks = artist_counts.iloc[0]

# Print the artist with the highest number of tracks and the count
print(f"The artist with the highest number of tracks is {artist_with_most_tracks} with {count_of_most_tracks} tracks.")

# Create a countplot
plt.figure(figsize=(10, 6))
sns.countplot(y='Artist', data=df, order=df['Artist'].value_counts().index)
plt.title('Number of Tracks per Artist')
plt.xlabel('Number of Tracks')
plt.ylabel('Artist')
plt.show()
```

```python
# Q5.What are the top 5 least popular tracks in the dataset? Provide the artist name and track name for each.


# Sort the DataFrame by popularity in ascending order and select the top 5 rows
least_popular_tracks = df.sort_values('Popularity').head(5)

# Display the artist name and track name for the least popular tracks
print(least_popular_tracks[['Artist', 'Track Name']])
```

```python
# Q6.Among the top 5 most popular artists, which artist has the highest popularity on average? Calculate and display the average popularity for each artist


# Group by artist and calculate the average popularity
artist_popularity = df.groupby('Artist')['Popularity'].mean()

# Sort the artists by average popularity in descending order and get the top 5
top_5_artists = artist_popularity.sort_values(ascending=False).head(5)

# Display the average popularity for each of the top 5 artists
print(top_5_artists)

# Find the artist with the highest average popularity among the top 5
highest_popularity_artist = top_5_artists.idxmax()
print(f"\nThe artist with the highest average popularity among the top 5 is: {highest_popularity_artist}")
```

```python
# Q7. For the top 5 most popular artists, what are their most popular tracks? List the track name for each artist.


# Group by artist and find the most popular track for each
top_tracks = df.loc[df.groupby('Artist')['Popularity'].idxmax()]

# Get the top 5 most popular artists
top_artists = df.groupby('Artist')['Popularity'].mean().sort_values(ascending=False).head(5).index

# Filter the top tracks to only include the top 5 artists
result = top_tracks[top_tracks['Artist'].isin(top_artists)]

# Display the track name for each artist
print(result[['Artist', 'Track Name']])
```

```python
# Q8. Visualize relationships between multiple numerical variables simultaneously using a pair plot.


# Select numerical columns for the pair plot
numerical_cols = ['Popularity', 'Duration (ms)']

# Create the pair plot
sns.pairplot(df[numerical_cols])
plt.show()
```

```python
# Q9.Does the duration of tracks vary significantly across different artists? Explore this visually using a box plot or violin plot


# Create the box plot
plt.figure(figsize=(12, 6))  # Adjust figure size for better readability
sns.boxplot(x='Artist', y='Duration (ms)', data=df)
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
plt.title('Distribution of Track Duration Across Artists')
plt.xlabel('Artist')
plt.ylabel('Duration (ms)')
plt.show()

```

```python
# Q10.How does the distribution of track popularity vary for different artists? Visualize this using a swarm plot or a violin plot


# Create the swarm plot
plt.figure(figsize=(12, 6))  # Adjust figure size as needed
sns.swarmplot(x="Artist", y="Popularity", data=df)
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
plt.title("Distribution of Track Popularity for Different Artists (Swarm Plot)")
plt.xlabel("Artist")
plt.ylabel("Popularity")
plt.show()
```
