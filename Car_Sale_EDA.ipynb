```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


import warnings
warnings.filterwarnings('ignore')

df = pd.read_csv('Car Sale.csv')
df
```

```python
df['year'] = df['Date'].str.split('/').str[2]
df['month'] = df['Date'].str.split('/').str[1]
df['day'] = df['Date'].str.split('/').str[0]

#df.drop('Date', axis=1, inplace=True)

df['year'] = df['year'].astype(int)
df['month'] = df['month'].astype(int)
df['day'] = df['day'].astype(int)
```

```python
df
```

```python
# 1. What is the average selling price of cars for each dealer, and how does it compare across different dealers?

average_price_by_dealer = df.groupby('Dealer_Name')['Price ($)'].mean()
average_price_by_dealer

```

```python
# Compare average prices across dealers (example: find the dealer with the highest average price)
highest_average_price = average_price_by_dealer.max()
dealer_with_highest_price = average_price_by_dealer[average_price_by_dealer == highest_average_price].index[0]

print(f"\nThe dealer with the highest average selling price is {dealer_with_highest_price} at ${highest_average_price:.2f}")
```

```python
# Q2.  Which car brand(Company) has the highest variation in prices, and what does this tell us about the pricing trends?

# Group data by car company and calculate the standard deviation of prices for each company
company_price_variation = df.groupby('Company')['Price ($)'].std()

# Find the company with the highest price variation
highest_variation_company = company_price_variation.idxmax()

# Print the result
print(f"The car brand with the highest price variation is: {highest_variation_company}")

# Analyze the pricing trends (example)
# A high standard deviation suggests a wide range of prices for that brand.
# This could be due to various factors like model year, features, or trim levels.
# Further investigation into the specific models and features within this brand is required to understand
# the reasons behind this price variation.
```

```python
# Q3.What is the distribution of car prices for each transmission type, and how do the interquartile ranges compare

# Group data by transmission type
grouped = df.groupby('Transmission')['Price ($)']

# Calculate quartiles for each group
quartiles = grouped.quantile([0.25, 0.5, 0.75])

# Calculate the interquartile range (IQR) for each group
iqr = quartiles.loc[0.75] - quartiles.loc[0.25]


# Create box plots to visualize distributions
plt.figure(figsize=(8, 6))
grouped.plot(kind='box')
plt.title('Distribution of Car Prices by Transmission Type')
plt.xlabel('Transmission Type')
plt.ylabel('Price ($)')


#Print IQR
print("Interquartile Ranges:")
print(iqr)

plt.show()
```

```python
# Q4. What is the distribution of car prices across different regions

# Group data by dealer region and get the mean price for each region
region_prices = df.groupby('Dealer_Region')['Price ($)'].mean()

# Create a bar plot to visualize the distribution of car prices across different regions
plt.figure(figsize=(10, 6))
sns.barplot(x=region_prices.index, y=region_prices.values)
plt.title('Distribution of Car Prices Across Different Regions')
plt.xlabel('Dealer Region')
plt.ylabel('Average Car Price')
plt.xticks(rotation=45, ha='right') # Rotate x-axis labels for better readability
plt.tight_layout()
plt.show()
```

```python
# Q5.What is the distribution of cars based on body styles?


# Calculate the distribution of cars based on body styles
body_style_counts = df['Body Style'].value_counts()

body_style_counts

plt.figure(figsize=(10, 6))
sns.barplot(x=body_style_counts.index, y=body_style_counts.values)
plt.title('Distribution of Cars Based on Body Styles')
plt.xlabel('Body Style')
plt.ylabel('Number of Cars')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# 6.How does the average selling price of cars vary by customer gender and annual income?

average_price_by_gender_income = df.groupby(['Gender', 'Annual Income'])['Price ($)'].mean().reset_index()

average_price_by_gender_income
```

```python
# Q7. What is the distribution of car prices by region, and how does the number of cars sold vary by region?


# Calculate the distribution of car prices by region
price_by_region = df.groupby('Dealer_Region')['Price ($)'].describe()

# Display the distribution
print("Distribution of car prices by region:\n", price_by_region)


# Calculate the number of cars sold by region
cars_sold_by_region = df.groupby('Dealer_Region')['Car_id'].count()

# Display the number of cars sold
print("\nNumber of cars sold by region:\n", cars_sold_by_region)


# Visualize the distribution of car prices by region using box plots
plt.figure(figsize=(12, 6))
sns.boxplot(x='Dealer_Region', y='Price ($)', data=df)
plt.title('Distribution of Car Prices by Region')
plt.xlabel('Region')
plt.ylabel('Price ($)')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.show()


# Visualize the number of cars sold by region using a bar plot
plt.figure(figsize=(12, 6))
cars_sold_by_region.plot(kind='bar')
plt.title('Number of Cars Sold by Region')
plt.xlabel('Region')
plt.ylabel('Number of Cars Sold')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.show()
```

```python
# Q8.  How does the average car price differ between cars with different engine sizes?


average_price_by_engine = df.groupby('Engine')['Price ($)'].mean()

average_price_by_engine

```

```python
# Q9. How do car prices vary based on the customer’s annual income bracket?


# Define income bracket bins and labels
income_bins = [0, 50000, 100000, 200000, float('inf')]
income_labels = ['Low', 'Medium', 'High', 'Very High']

# Bin the Annual Income
df['Income Bracket'] = pd.cut(df['Annual Income'], bins=income_bins, labels=income_labels, right=False)

# Calculate the mean price for each income bracket
price_by_income = df.groupby('Income Bracket')['Price ($)'].mean()

# Print or visualize the result
print(price_by_income)

# Optional: Create a bar plot

plt.figure(figsize=(8, 6))
price_by_income.plot(kind='bar')
plt.title('Average Car Price by Annual Income Bracket')
plt.xlabel('Income Bracket')
plt.ylabel('Average Car Price ($)')
plt.xticks(rotation=0)
plt.show()

```

```python
# Q10. What are the top 5 car models with the highest number of sales, and how does their price distribution look?

# Find the top 5 car models by sales
top_5_models = df['Model'].value_counts().nlargest(5).index

# Create a plot of the price distribution for each top model


plt.figure(figsize=(12, 6))
for model in top_5_models:
    sns.kdeplot(df[df['Model'] == model]['Price ($)'], label=model, fill=True)

plt.title('Price Distribution of Top 5 Car Models')
plt.xlabel('Price')
plt.ylabel('Density')
plt.legend()
plt.show()
```

```python
#Q11. How does car price vary with engine size across different car colors, and which colors have the highest price variation


# Group data by engine size and color, then calculate the mean price
price_variation = df.groupby(['Engine', 'Color'])['Price ($)'].mean().unstack()

# Plot the price variation
price_variation.plot(kind='bar')
plt.title('Car Price Variation by Engine Size and Color')
plt.xlabel('Engine Size')
plt.ylabel('Average Price ($)')
plt.xticks(rotation=45)
plt.legend(title='Color')
plt.tight_layout()
plt.show()

# Calculate the price range for each color
price_range_by_color = df.groupby('Color')['Price ($)'].agg(['min', 'max'])

# Calculate price variation for each color
price_range_by_color['variation'] = price_range_by_color['max'] - price_range_by_color['min']

# Find colors with the highest price variation
highest_variation_colors = price_range_by_color.sort_values('variation', ascending=False)

# Print the result
print("Colors with highest price variation:")
highest_variation_colors
```

```python
# Q12.Is there any seasonal trend in car sales based on the date of sale?

# Convert the 'Date' column to datetime objects
df['Date'] = pd.to_datetime(df['Date'])

# Extract month and year from the 'Date' column
df['Month'] = df['Date'].dt.month
df['Year'] = df['Date'].dt.year

# Group sales by month and year
monthly_sales = df.groupby(['Year', 'Month'])['Car_id'].count().reset_index()

# Plot the seasonal trend
plt.figure(figsize=(12, 6))
plt.plot(monthly_sales['Month'], monthly_sales['Car_id'])
plt.xlabel('Month')
plt.ylabel('Number of Cars Sold')
plt.title('Seasonal Trend in Car Sales')
plt.grid(True)
plt.show()
```

```python
# Q13. How does the car price distribution change when considering different combinations of body style and transmission type

# Group data by 'Body Style' and 'Transmission' and get the price distribution.
grouped_data = df.groupby(['Body Style', 'Transmission'])['Price ($)']

# Plot distributions for each combination
plt.figure(figsize=(15, 10))

for (body_style, transmission), prices in grouped_data:
    sns.histplot(prices, kde=True, label=f"{body_style} - {transmission}")

plt.title("Car Price Distribution by Body Style and Transmission Type")
plt.xlabel("Price ($)")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```

```python
# Q14.What is the correlation between car price, engine size, and annual income of customers, and how do these features interact?

# Calculate correlations
correlation_matrix = df[['Price ($)', 'Annual Income']].corr()

# Print the correlation matrix
print("Correlation Matrix:")
print(correlation_matrix)


# Analyze interactions (example: using scatter plot)
# Import plotting libraries
import matplotlib.pyplot as plt
import seaborn as sns


# Create a scatterplot to visualize the interaction
sns.scatterplot(x='Annual Income', y='Price ($)', hue='Engine', data=df)
plt.title('Interaction between Annual Income, Price, and Engine Type')
plt.show()
```
