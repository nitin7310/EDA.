```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


import warnings
warnings.filterwarnings('ignore')

df = pd.read_csv('amazon.csv')
df
```

```python
# Q1. What is the average rating for each product category?

# Convert 'rating' column to numeric, handling errors
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')

# Group by 'category' and calculate the mean rating
average_rating_by_category = df.groupby('category')['rating'].mean()

# Display the result
average_rating_by_category


```

```python
# Q2. What are the top rating_count products by category?

df['rating_count'] = df['rating_count'].astype(str).str.replace(',', '').str.replace('[\u20b9,]', '', regex=True)
df['rating_count'] = pd.to_numeric(df['rating_count'], errors='coerce')

# Group by category and get the top rating_count for each category
top_products = df.groupby('category')['rating_count'].idxmax()

# Display the top products
print(df.loc[top_products, ['category', 'product_name', 'rating_count']])
```

```python
# Q3. What is the distribution of discounted prices vs. actual prices?

# Convert 'discounted_price' and 'actual_price' to numeric, handling errors
df['discounted_price'] = df['discounted_price'].astype(str).str.replace(r'[^\d.]', '', regex=True)
df['actual_price'] = df['actual_price'].astype(str).str.replace(r'[^\d.]', '', regex=True)

# Convert to numeric, coercing errors to NaN
df['discounted_price'] = pd.to_numeric(df['discounted_price'], errors='coerce')
df['actual_price'] = pd.to_numeric(df['actual_price'], errors='coerce')

# Drop rows with NaN values in 'discounted_price' and 'actual_price' columns
df.dropna(subset=['discounted_price', 'actual_price'], inplace=True)


# Create the distribution plot
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='discounted_price', label='Discounted Price', kde=True, color="skyblue")
sns.histplot(data=df, x='actual_price', label='Actual Price', kde=True, color="salmon")

# Customize the plot
plt.title('Distribution of Discounted Prices vs. Actual Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.legend()
plt.show()

```

```python
# Q4.How does the average discount percentage vary across categories?

# Convert 'discount_percentage' column to numeric, handling errors
df['discount_percentage'] = pd.to_numeric(df['discount_percentage'].str.rstrip('%'), errors='coerce')

# Group data by category and calculate the average discount percentage
average_discount_by_category = df.groupby('category')['discount_percentage'].mean()

# Create a bar plot to visualize the average discount percentage across categories
plt.figure(figsize=(10, 6))  # Adjust figure size for better visualization
average_discount_by_category.plot(kind='bar')
plt.title('Average Discount Percentage by Category')
plt.xlabel('Category')
plt.ylabel('Average Discount Percentage')
plt.xticks(rotation=90)  # Rotate x-axis labels for readability
plt.tight_layout()  # Adjust layout to prevent labels from overlapping
plt.show()
```

```python
# Q5. What are the most popular product names?

# Group by product name and count the occurrences
product_counts = df.groupby('product_name').size().reset_index(name='counts')

# Sort the product names by counts in descending order
product_counts = product_counts.sort_values('counts', ascending=False)

# Display the most popular product names
print(product_counts.head(10)) # Display the top 10, change the number as needed
```

```python
# Q6. What are the most popular product keywords?

# Combine all text columns into a single column for keyword extraction
text_columns = ['product_name', 'category', 'about_product', 'review_title', 'review_content']
df['all_text'] = df[text_columns].apply(lambda row: ' '.join(row.values.astype(str)), axis=1)

# Function to extract keywords
def extract_keywords(text):
    # Convert text to lowercase
    text = text.lower()
    # Remove punctuation and special characters (you might need to customize this based on your needs)
    text = ''.join(c for c in text if c.isalnum() or c.isspace())
    # Tokenize the text into words
    words = text.split()
    return words

# Extract keywords from all text data
all_keywords = []
for text in df['all_text']:
    all_keywords.extend(extract_keywords(text))

# Count the frequency of each keyword
word_counts = Count(all_keywords)

# Get the top 10 most frequent keywords
top_keywords = word_counts.most_common(10)

# Display the results
print("Top 10 most popular product keywords:")
for keyword, count in top_keywords:
    print(f"- {keyword}: {count}")
```

```python
# Q7. What are the most popular product reviews?

# Calculate the average rating for each product
average_ratings = df.groupby('product_name')['rating'].mean()

# Sort the products by average rating in descending order
sorted_ratings = average_ratings.sort_values(ascending=False)

# Display the top 10 most popular products based on average rating
print(sorted_ratings.head(10))
```

```python
# Q8.What is the correlation between discounted_price and rating?

# Calculate the correlation between 'discounted_price' and 'rating'.
correlation = df['discounted_price'].corr(df['rating'])

# Print the correlation coefficient.
print(f"The correlation between discounted_price and rating is: {correlation}")
```

```python
# Q9. What are the Top 5 categories based on the highest ratings?

# Group by category and calculate the mean rating for each category
category_ratings = df.groupby('category')['rating'].mean()

# Sort the categories by mean rating in descending order and get the top 5
top_5_categories = category_ratings.sort_values(ascending=False).head(5)

# Display the top 5 categories
top_5_categories
```
